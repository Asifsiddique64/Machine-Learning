```python
import statsmodels.api as sm
```

```python
int_rate=data['Interest.Rate']
loan_amt=data['Loan.Amount']
fico=data['FICO.Score']
```

```python
y=np.matrix(int_rate).transpose()
x1=np.matrix(fico).transpose()
x2=np.matrix(loan_amt).transpose()
```

```python
x=np.column_stack([x1,x2])
```

```python
x3=sm.add_constant(x)
```

```python
#OLS ordinary least square
```

```python
model=sm.OLS(y,x3)
model_fit=model.fit()
```

```python
print ("the p values are",model_fit.pvalues) # this value should be less than 0.05
print ("print the R-squared value is",model_fit.rsquared) #how well the regress line appr in data and its value lies between -1 to 1 and the closer it is to 1 the better
```
